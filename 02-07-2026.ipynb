```python
# ===============================
#   University of Mines and Technology Management System
# ===============================

# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


# Child Class 1 — Student
class Student(Person):
    def __init__(self, name, age, student_id, enroll_course):
        super().__init__(name, age)        # Inherits from Person
        self.student_id = student_id
        self.enroll_course = enroll_course

    def enroll(self):
        print(f"{self.name} has enrolled in {self.enroll_course}")

    # Override display_info to include student ID
    def display_info(self):
        super().display_info()             # Calls Person's display_info
        print(f"Student ID: {self.student_id}")
        print(f"Course: {self.enroll_course}")


# Child Class 2 — Lecturer
class Lecturer(Person):
    def __init__(self, name, age, employee_id, teach_course):
        super().__init__(name, age)        # Inherits from Person
        self.employee_id = employee_id
        self.teach_course = teach_course

    def teach(self):
        print(f"{self.name} is teaching {self.teach_course}")

    # Override display_info to include employee ID
    def display_info(self):
        super().display_info()             # Calls Person's display_info
        print(f"Employee ID: {self.employee_id}")
        print(f"Teaching: {self.teach_course}")


# ===============================
#        Create Objects
# ===============================
Gyan = Student("Gyan", 20, "FOE.41.006.088.25",
                  "Electrical and Electronic Engineering")

Matthew = Lecturer("Matthew", 39, "LE_27",
                   "Object Oriented Programming")

# ===============================
#        Test Everything
# ===============================

print("=" * 40)
print("        STUDENT INFO")
print("=" * 40)
Gyan.display_info()         # Inherited + overridden
Gyan.enroll()               # Student's own method

print()
print("=" * 40)
print("        LECTURER INFO")
print("=" * 40)
Matthew.display_info()         # Inherited + overridden
Matthew.teach()                # Lecturer's own method
```
